# Dataset 

### Daily Exchange Rates per Euro 1999-2021
Source: [kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020 "Daily Exchange Rates per Euro 1999-2021")

Reference rates are euro foreign exchange rates observed on major foreign exchange trading venues at a certain point in time = they are the price of one currency in terms of another currency. The rates are usually updated around 16:00 CET on every working day, except on TARGET closing days.

Dataset contains date and Euro rate corresponding to Australian dollar, Bulgarian lev, Brazilian real, Canadian dollar, Swiss franc, Chinese yuan renminbi, Cypriot pound, Czech koruna, Danish krone, Estonian kroon, UK pound sterling, Greek drachma, Hong Kong dollar, Croatian kuna, Hungarian forint, Indonesian rupiah, Israeli shekel, Indian rupee, Iceland krona, Japanese yen, Korean won, Lithuanian litas, Latvian lats, Maltese lira, Mexican peso, Malaysian ringgit, Norwegian krone, New Zealand dollar, Philippine peso, Polish zloty, Romanian leu, Russian rouble, Swedish krona, Singapore dollar, Slovenian tolar, Slovak koruna, Thai baht, Turkish lira, US dollar, South African rand.

- Some currency in the list doesn't exist anymore; it was replaced by the Euro €: Cypriot pound (2007), Estonian kroon (2011), Greek drachma (2002), Lithuanian litas (2015), Latvian lats (2014), Maltese lira (2008), Slovenian tolar (2007), Slovak koruna (2009). 
- Bulgarian lev since 2002 is pegged to the Euro: 1 € = 1.9558 leva.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')

exchange_rates.head()
```

# Cleaning and Isolating the Data

We're only interested in visualising one currency in this exercise.

Therefore, we can:
- isolate that currency
- clean the reduced dataset

```python
# Rename the two columns we're interested in to something easier to read and type
exchange_rates.rename(columns={'[US dollar ]': 'us_dollar','[UK pound sterling ]': 'uk_pound', 'Period\\Unit:': 'time'}, inplace=True)

# Cast the time column as datetime
exchange_rates['time'] = pd.to_datetime(exchange_rates['time'])

# Sort by time ascending and create a new index
exchange_rates.sort_values('time', ignore_index=True, inplace=True)

# Isolate the columns we're going to visualise
euro_to_pound = exchange_rates[['time', 'uk_pound']]

# Remove rows with - in them insted of a conversation rate
euro_to_pound = euro_to_pound[euro_to_pound['uk_pound'] != '-']

# Case uk_pound as float
euro_to_pound['uk_pound'] = euro_to_pound['uk_pound'].astype(float)
```

```python
euro_to_pound.info()
```

```python
plt.plot(euro_to_pound['time'], euro_to_pound['uk_pound'])
plt.show()
```

```python
# Calculate the rolling means for uk_pound on a 30 day window
euro_to_pound['rolling_30'] = euro_to_pound['uk_pound'].rolling(30).mean()

plt.plot(euro_to_pound['time'], euro_to_pound['rolling_30'])
plt.show()
```

# Visualisation

## Message
The visualisation will show **The Impact of Prime Ministers on Euro Exchange Rates**

## Preparing the Data for Charting

### Slices will be made for:
- Boris Johnson: 06 2019 to Present
- Theresa May  : 06 2016 to 06 2019
- David Cameron: 06 2010 to 06 2016
- Labour : 01 1991 to 06 2010

### Key Dates
- October 2008, RBS Bailed Out
- December 2014, Export Gains
- June 2016, Brexit Vote
- December 2020, Transition Period Ends



```python
# Visualisation Layer Slices

# Boris Johnson's Term
johnson_slice = euro_to_pound.set_index('time')['2019-06-15':]
johnson_slice.reset_index(inplace=True)

# Theresa May's Term
may_slice = euro_to_pound.set_index('time')['2016-06-15':'2019-06-15']
may_slice.reset_index(inplace=True)

# David Cameron's Term
cameron_slice = euro_to_pound.set_index('time')['2010-06-15':'2016-06-15']
cameron_slice.reset_index(inplace=True)

# Labour's Term
labour_slice = euro_to_pound.set_index('time')[:'2010-06-15']
labour_slice.reset_index(inplace=True)

# Whole Chart
chart_slice = euro_to_pound.set_index('time')
chart_slice.reset_index(inplace=True)

# Key Dates

# October 2008, RBS Bailed Out
lb_index = int(chart_slice[chart_slice['time'] == '2008-10-15'].index.values) # Get the index value of the date we want as an integer.
lb_date = chart_slice['time'].iloc[lb_index] # Use the index to get the long string date for charting.

# December 2014, Export Gains
eg_index = int(chart_slice[chart_slice['time'] == '2014-12-15'].index.values) # Get the index value of the date we want as an integer.
eg_date = chart_slice['time'].iloc[eg_index] # Use the index to get the long string date for charting.

# June 2016, Brexit Vote
bv_index = int(chart_slice[chart_slice['time'] == '2016-06-15'].index.values) # Get the index value of the date we want as an integer.
bv_date = chart_slice['time'].iloc[bv_index] # Use the index to get the long string date for charting.

# December 2020, Transition Period Ends
tp_index = int(chart_slice[chart_slice['time'] == '2020-12-15'].index.values) # Get the index value of the date we want as an integer.
tp_date = chart_slice['time'].iloc[tp_index] # Use the index to get the long string date for charting.
```

## Charting the Data

### Method
The following Gestalt principles will be use:
- Continuation
- Similarity
to create a **journey** for the viewer.

### Design
- 4 charts showing 4 periods of Euro exchange rates under Labour + 3 different Tory PMs.
- 1 base chart is repeated 4 times.
- A chart is layered on top for each of the periods above.
- 4 key events in UK economics and domestic policy are indicated, one on each graph

```python
#Style
style.use('default')

fig,(ax1,ax2,ax3,ax4) = plt.subplots(ncols=1, nrows=4, figsize=(7,8))

# Useful lists
axes = [ax1,ax2,ax3,ax4]
sides = ['left','bottom','right','top']

# Calculate the historic trend line, in this instance the historic mean of the Euro to Pound.
historic_mean = chart_slice['uk_pound'].mean()

# Loop to set the attributes
y_pos = 0
for ax in axes:
    # Baseline charts for layering on top of.
    ax.plot(chart_slice['time'], chart_slice['rolling_30'], alpha=0.3, color='grey')
    
    # Historic trend line
    ax.axhline(y=historic_mean, alpha=0.5, lw=1, color='grey')
    
    # Style
    ax.grid(b=False)
    ax.set(facecolor='white')
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.tick_params(bottom=0, left=0) 
    for side in sides:
        ax.spines[side].set_visible(False)

# Layer slices over the baseline chart
ax1.plot(labour_slice['time'],labour_slice['rolling_30'], alpha=1,color='#DC241f')
ax2.plot(cameron_slice['time'],cameron_slice['rolling_30'], alpha=1,color='#0087DC')
ax3.plot(may_slice['time'],may_slice['rolling_30'], alpha=1,color='#0087DC')
ax4.plot(johnson_slice['time'],johnson_slice['rolling_30'], alpha=1,color='#0087DC')

# Title and subtitle
ax1.text(0,1.3, s='The Impact of Prime Ministers on Euro Exchange Rates', color='black', weight='bold', size='13', transform=ax1.transAxes)
ax1.text(0,1.15, s='Source: European Central Bank, 1999 to Present', color='black', size='11', transform=ax1.transAxes)

# Annotate with text
ax1.axvline(x=lb_date, color='#78c247', lw=1, ymin=0.2)
ax1.text(0.4,-0.03, s='Bailout\nof RBS', color='#78c247',transform=ax1.transAxes, weight='bold', multialignment='center')

ax2.axvline(x=eg_date, color='#78c247', lw=1, ymin=0.2)
ax2.text(.64,-1.2, s='Exports\nBoom', color='#78c247',transform=ax1.transAxes, weight='bold',multialignment='center')

ax3.axvline(x=bv_date, color='#78c247', lw=1, ymin=0.2)
ax3.text(.72,-2.43, s='Brexit\nVote', color='#78c247', weight='bold',transform=ax1.transAxes, multialignment='center')

ax4.axvline(x=tp_date, color='#78c247', lw=1, ymin=0.2)
ax4.text(.89,-3.63, s='UK Exits\nthe EU', color='#78c247', transform=ax1.transAxes, weight='bold',multialignment='center')

# Footer
ax4.text(0,-3.85, s='Data Visualisation by James Holiday-Scott, 2021'+ ' '*52, color='white', transform=ax1.transAxes, backgroundcolor='grey')

plt.show()
```

# Summary
**Gestalt principles** of design - continuation and similarity - create a feeling of movement against a historic trend.

**Intentionally emotive**, and certainly not good journalism, the descending line of continuation adds a feeling of decline to the chart which is 100% intentional.

**Data:Ink** ratio is very high, making use of only the shape of the exchange rate chart and the historical trend line of the mean exchange rate to give numerical relation.
